1. What is a random variable in probability theory
-   A  random variable is a variable that takes on different values
     based on the outcome of a random event.
-   It is a function that assigns numerical values to each possible
     outcome in a sample space.


2. What are the types of random variables
-  Types of Random Variables
   1. Discrete Random Variable

           Takes countable values (eg:- 0, 1, 2, ...)

           Example: The number of heads in 3 coin flips (0, 1, 2, or 3).

   2. Continuous Random Variable

          Takes infinite, uncountable values within a range.

          Example: The time it takes for a bus to arrive (eg:- 5.3 minutes, 6.7 minutes).



3. What is the difference between discrete and continuous distributions
-   Discrete distribution:-
  -  countable outcomes
  -  Discrete random variable
  -  Uses probability mass function (PMF) for calculating probability.
  - Example:- Number of heads in 10 coin flips (0, 1, 2, …, 10)

-  Continuous distribution:-
   - uncountable outcomes within a range
   - Continuous random variable
   - Uses probability density function (PDF) for calculating probability.
   -  Example:- Heights of students (e.g., 5.6 ft, 5.8 ft)



4. What are probability distribution functions (PDF)
-   A Probability Distribution Function (PDF) describes how probabilities are distributed over the values of a random variable. - -
-   There are two main types:

       1. Probability Mass Function (PMF) – for discrete random variables.
       2. Probability Density Function (PDF) – for continuous random variables.



5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)
-   Cumulative Distribution Function (CDF):-
    - Represents the probability that a random variable X is less than or equal to a given value x.
   - Always increases from 0 to 1.

-   Probability Distribution Function (PDF):-
    - Describes how probability is distributed over a range of values.
    - Used for both discrete (PMF) and continuous (PDF) random variables.
    - In the continuous case, the PDF does not give probabilities directly but rather density values.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics
-  Central limit Theorem state that we have a population with mean μ
 and standard deiation σ  and take sufficiently large number of random sample from the population with replacement , then the distribution of sample means will be approximately normally distributed.
- The Central Limit Theorem (CLT) is crucial in statistics because it allows us to make inferences about populations using sample data, even when the original population distribution is unknown or not normal.

14.  How does the Central Limit Theorem relate to the normal distribution
-   The Central Limit Theorem (CLT) is directly connected to the normal distribution because it explains why normal distributions appear so frequently in statistics and allows us to use normal-based statistical methods in many real-world scenarios.



15.  What is the application of Z statistics in hypothesis testing
-  Z-statistics are widely used in hypothesis testing when the population standard deviation is known or the sample size is large (n≥30). Z-tests help determine whether sample data provides enough evidence to reject a null hypothesis.



16. How do you calculate a Z-score, and what does it represent

          Z= ( X - μ )/ σ

17.  What are point estimates and interval estimates in statistics
-     Point Estimates:-
           A point estimate is a single value used to estimate an unknown population parameter.
-    Interval Estimates
          An interval estimate gives a range of values (instead of a single point) within which the population parameter is likely to lie.

18.  What is the significance of confidence intervals in statistical analysis
-   In statistical analysis, confidence intervals are crucial for estimating population parameters based on sample data. They provide a range of values within which the true population parameter is likely to lie, offering insights into the precision and reliability of the estimate.

19.  What is the relationship between a Z-score and a confidence interval
-  A higher confidence level (e.g. 99%) results in a larger Z-score, making the confidence interval wider.

-  A lower confidence level (e.g., 90%) has a smaller Z-score, leading to a narrower confidence interval.

20.  How are Z-scores used to compare different distributions
-  Imagine two students take different standardized tests:
rani takes the SAT and scores 1300 (mean = 1050, standard deviation = 150).

-   Boby takes the ACT and scores 30 (mean = 21, standard deviation=5).

  rani's  Z = (1300-1050)/150 = 250/150 = 1.67
  bobs z = (30-21)/5 = 9/5 = 1.8
  -    Boby Z-score (1.8) is higher than rani (1.67).


21.  What are the assumptions for applying the Central Limit Theorem
-   1. Random Sampling
           The sample must be selected randomly from the population.
           Each observation should be independent (one selection should not influence another).
           If sampling without replacement, the population should be large compared to the sample size (usually N≥10n).

    2. Sample Size is Large Enough
           A general rule is that n≥30 is usually sufficient for the CLT to apply.

           If the population distribution is heavily skewed or non-normal, a larger sample size may be required (n>50 or more).
           If the population itself is normally distributed, even small samples (e.g., n<30) will have a normally distributed sampling mean.

    3. Finite Variance
           The population must have a finite mean and variance.
           If the variance is infinite (e.g., Cauchy distribution), the CLT does not hold.



22.  What is the concept of expected value in a probability distribution
-   The expected value (EV) of a probability distribution represents the long-run average outcome of a random variable over many trials. It gives the theoretical mean of a distribution.

23.  How does a probability distribution relate to the expected outcome of a random variable?
-   A probability distribution specifies the likelihood of each possible outcome for a random variable, and the expected value is the weighted average of these outcomes, representing the long-term average result over numerous trials.

```python
# Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

# Parameters
num_samples = 1000
num_iterations = 10000

def central_limit_theorem():
    original_data = np.random.exponential(scale=2.0, size=num_samples)
    #generating mean
    sample_means = [np.mean(np.random.choice(original_data, num_samples, replace=True)) for _ in range(num_iterations)]

    #Ploting histogram of sample means
    plt.figure(figsize=(8, 5))
    plt.hist(sample_means, bins=50, density=True, alpha=0.7, color='blue', edgecolor='black')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.title('Central Limit Theorem Demonstration')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# Run CLT simulation
central_limit_theorem()
```

```python
# Implement hypothesis testing using Z-statistics for a sample dataset
def hypothesis_test(sample_mean, population_mean, population_std, sample_size, alpha=0.05):
    # Calculating Z-statistic
    standard_error = population_std / np.sqrt(sample_size)
    z_stat = (sample_mean - population_mean) / standard_error

    # Calculating p-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    # Determining result of hypothesis test
    reject_null = p_value < alpha
    return z_stat, p_value, reject_null


# Parameters for hypothesis test
population_mean = 50
population_std = 10
sample_size = 30
alpha = 0.05

# Generating sample data from normal distribution
sample_data = np.random.normal(population_mean, population_std, sample_size)
sample_mean = np.mean(sample_data)

# Perform hypothesis test
z_stat, p_value, reject_null = hypothesis_test(sample_mean, population_mean, population_std, sample_size, alpha)

# Generate values for standard normal distribution
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

# Ploting the standard normal distribution with Z-statistic
plt.figure(figsize=(8, 5))
plt.plot(x, y, color='blue', label='Standard Normal Distribution')
plt.axvline(z_stat, color='red', linestyle='dashed', label=f'Z-stat = {z_stat:.2f}')
plt.xlabel('Z-score')
plt.ylabel('Density')
plt.title('Hypothesis Testing using Z-statistics')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Printing the results
print(f'Z-statistic: {z_stat:.2f}')
print(f'P-value: {p_value:.4f}')
print(f'Reject Null Hypothesis: {reject_null}')
```

```python
# Create a confidence interval for a dataset using Python and interpret the result

def confidence_interval(data, confidence=0.95):
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    sample_size = len(data)

    # Calculating margin of error
    z_score = norm.ppf((1 + confidence) / 2)
    margin_of_error = z_score * (sample_std / np.sqrt(sample_size))

    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound, sample_mean


# Parameters
population_mean = 50
population_std = 10
sample_size = 30
confidence = 0.95

# Generating sample data from normal distribution
sample_data = np.random.normal(population_mean, population_std, sample_size)

# Computing confidence interval
lower, upper, sample_mean = confidence_interval(sample_data, confidence)

# Generating values for normal distribution
x = np.linspace(sample_mean - 4, sample_mean + 4, 1000)
y = norm.pdf(x, sample_mean, population_std / np.sqrt(sample_size))

# Ploting
plt.figure(figsize=(8, 5))
plt.plot(x, y, color='blue', label='Sample Distribution')
plt.axvline(lower, color='red', linestyle='dashed', label=f'Lower Bound: {lower:.2f}')
plt.axvline(upper, color='green', linestyle='dashed', label=f'Upper Bound: {upper:.2f}')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.title(f'{confidence*100:.0f}% Confidence Interval')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Printing results
print(f'Sample Mean: {sample_mean:.2f}')
print(f'{confidence*100:.0f}% Confidence Interval: ({lower:.2f}, {upper:.2f})')
```

```python
# Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean

def confidence_interval(data, confidence=0.95):
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    sample_size = len(data)

    # Calculating margin of error
    z_score = norm.ppf((1 + confidence) / 2)
    margin_of_error = z_score * (sample_std / np.sqrt(sample_size))

    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound, sample_mean


# Parameters
population_mean = 50
population_std = 10
sample_size = 30
confidence = 0.95

# Generating sample data from normal distribution
sample_data = np.random.normal(population_mean, population_std, sample_size)

# Computing confidence interval
lower, upper, sample_mean = confidence_interval(sample_data, confidence)

# Generating values for normal distribution curve
x = np.linspace(sample_mean - 4 * population_std / np.sqrt(sample_size),
                 sample_mean + 4 * population_std / np.sqrt(sample_size), 1000)
y = norm.pdf(x, sample_mean, population_std / np.sqrt(sample_size))

# Ploting confidence interval
plt.figure(figsize=(8, 5))
plt.plot(x, y, color='blue', label='Sample Distribution')
plt.axvline(lower, color='red', linestyle='dashed', label=f'Lower Bound: {lower:.2f}')
plt.axvline(upper, color='green', linestyle='dashed', label=f'Upper Bound: {upper:.2f}')
plt.axvline(sample_mean, color='black', linestyle='dotted', label=f'Sample Mean: {sample_mean:.2f}')
plt.fill_between(x, y, where=(x >= lower) & (x <= upper), color='gray', alpha=0.3, label='Confidence Interval')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.title(f'{confidence*100:.0f}% Confidence Interval for Normal Distribution')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Printing results
print(f'Sample Mean: {sample_mean:.2f}')
print(f'{confidence*100:.0f}% Confidence Interval: ({lower:.2f}, {upper:.2f})')
```

```python
# Write a Python program to compare the standard deviations of two datasets and visualize the difference

# Generating two datasets
size = 1000
np.random.seed(42)
data1 = np.random.normal(50, 10, size)
data2 = np.random.normal(50, 20, size)

# Calculating standard deviations
std1 = np.std(data1, ddof=1)
std2 = np.std(data2, ddof=1)

# Ploting histograms
plt.figure(figsize=(8, 5))
plt.hist(data1, bins=30, alpha=0.6, color='blue', label=f'Dataset 1 (Std={std1:.2f})')
plt.hist(data2, bins=30, alpha=0.6, color='orange', label=f'Dataset 2 (Std={std2:.2f})')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Comparison of Standard Deviations')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Printng  standard deviations
print(f'Standard Deviation of Dataset 1: {std1:.2f}')
print(f'Standard Deviation of Dataset 2: {std2:.2f}')
```

```python
# Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution
# Generating dataset
size = 1000
np.random.seed(42)
data = np.random.normal(50, 15, size)  # Mean=50, Std=15

# Calculating range and interquartile range
data_range = np.max(data) - np.min(data)
q1, q3 = np.percentile(data, [25, 75])
iqr = q3 - q1

# Ploting histogram
plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, alpha=0.7, color='blue', edgecolor='black')
plt.axvline(q1, color='red', linestyle='dashed', label=f'Q1: {q1:.2f}')
plt.axvline(q3, color='green', linestyle='dashed', label=f'Q3: {q3:.2f}')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Dataset with IQR')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Printing range and IQR
print(f'Range of Dataset: {data_range:.2f}')
print(f'Interquartile Range (IQR): {iqr:.2f}')
```

```python
# Implement Z-score normalization on a dataset and visualize its transformation

# Generating dataset
size = 1000
np.random.seed(42)
data = np.random.normal(50, 15, size)
# Performing Z-score normalization
mean = np.mean(data)
std = np.std(data, ddof=1)
z_normalized_data = (data - mean) / std

# Ploting histograms before and after normalization
plt.figure(figsize=(12, 5))

# Original data histogram
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, alpha=0.7, color='blue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Original Dataset')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Z-score normalized data histogram
plt.subplot(1, 2, 2)
plt.hist(z_normalized_data, bins=30, alpha=0.7, color='green', edgecolor='black')
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.title('Z-score Normalized Dataset')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()

# Print mean and standard deviation before and after normalization
print(f'Original Mean: {mean:.2f}, Original Std Dev: {std:.2f}')
print(f'Normalized Mean: {np.mean(z_normalized_data):.2f}, Normalized Std Dev: {np.std(z_normalized_data, ddof=1):.2f}')
```

```python
# Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution

from scipy.stats import skew, kurtosis

# Generating dataset
size = 1000
np.random.seed(42)
data = np.random.normal(50, 15, size)

# Calculating skewness and kurtosis
data_skewness = skew(data)
data_kurtosis = kurtosis(data)

# Ploting histogram
plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, alpha=0.7, color='blue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Dataset')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Print skewness and kurtosis
print(f'Skewness of Dataset: {data_skewness:.2f}')
print(f'Kurtosis of Dataset: {data_kurtosis:.2f}')
```
